## Turn the labels to one-hot encording

```python
one_hot = df.labels.str.split('\s*,\s*', expand=True) \
   .stack() \
   .str.get_dummies() \
   .sum(level=0)
one_hot
```

```python
df = df.drop(columns='labels').join(one_hot, how='outer')
```
